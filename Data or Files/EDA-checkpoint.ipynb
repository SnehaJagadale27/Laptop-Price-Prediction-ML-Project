```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv("Data_clean.csv")
ds
```

```python
ds.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
ds
```

```python
ds.info()
```

```python
#Distribution Of Car Prices
plt.figure(figsize=(10,5))
sns.histplot(ds['Price (₹)'],bins=30,kde=True,color='skyblue')
plt.title('Distribution Of Laptop Prices')
plt.xlabel('(Price (₹))')
plt.ylabel('Frequency')
plt.show()
```

```python

# Price Comparison of Top Brands
top5=ds['Brand'].value_counts().head(5).index
plt.figure(figsize=(10,5))
sns.boxplot(data=ds[ds['Brand'].isin(top5)],x='Brand',y='Price (₹)',palette='pastel')
plt.title('Price Distribution of Top 5 Laptop Brands')
plt.xlabel('Brand')
plt.ylabel('Price (₹)')
plt.show()
```

```python
# Brand Type Distribution

brand_co=ds['Brand'].value_counts()
plt.figure(figsize=(10,5))
plt.pie(brand_co,labels=brand_co.index,autopct='%1.1f%%',startangle=140,colors=sns.color_palette('pastel'))
plt.title('Brand type Distribution')
plt.show()
```

```python
#GPU Distribution
plt.figure(figsize=(8,5))
plt.title('GPU Distribution')
sns.barplot(x=ds['GPU'],y=ds['Price (₹)'])
ds['GPU'].value_counts().plot(kind='bar')
# sns.scatterplot(x=ds['Screen Size (inches)'],y=ds['Price (₹)'])
```
